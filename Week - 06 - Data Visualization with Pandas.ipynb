```python
import pandas as pd  
import matplotlib.pyplot as plt  

# Creating DataFrame
df = pd.DataFrame({  
    'Category': ['A', 'B', 'C', 'D'],  
    'Values': [4, 7, 1, 8]  
})  

# Plotting Bar Plot
df.plot(kind='bar', x='Category', y='Values', color='purple')  
plt.title("Bar Plot from Pandas DataFrame")  
plt.xlabel("Category")  
plt.ylabel("Values")  
plt.show()  
```

```python
import pandas as pd  
import matplotlib.pyplot as plt  

# Creating DataFrame
df = pd.DataFrame({  
    'X': [1, 2, 3, 4],  
    'Y': [10, 20, 25, 30]  
})  

# Plotting Line Plot
df.plot(kind='line', x='X', y='Y', marker='o')  
plt.title("Line Plot from Pandas DataFrame")  
plt.xlabel("X-axis")  
plt.ylabel("Y-axis")  
plt.show()  
```

```python
import pandas as pd  
import matplotlib.pyplot as plt  

# Creating DataFrame
df = pd.DataFrame({  
    'Categories': ['A', 'B', 'C', 'D'],  
    'Values': [20, 30, 25, 25]  
})  

# Plotting Pie Chart
df.plot.pie(y='Values', labels=df['Categories'], autopct='%1.1f%%', startangle=90)  
plt.title("Pie Chart from Pandas DataFrame")  
plt.ylabel('')  
plt.show()  
```
