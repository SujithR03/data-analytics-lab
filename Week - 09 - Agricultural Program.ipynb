```python
import pandas as pd

# Load the agricultural dataset
df = pd.read_csv('agriculture_data.csv')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Display summary statistics of the dataset
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize crop yield distribution by crop type using a boxplot
sns.boxplot(x='crop_type', y='yield', data=df)
plt.title("Crop Yield Distribution by Type")
plt.xlabel("Crop Type")
plt.ylabel("Yield")
plt.show()
```
