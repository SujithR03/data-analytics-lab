```python
import matplotlib.pyplot as plt  

# Data for line plot
x = [1, 2, 3, 4, 5]  
y = [2, 3, 5, 7, 11]  

# Plotting the line
plt.plot(x, y)  
plt.title("Line Plot")  
plt.xlabel("X-axis")  
plt.ylabel("Y-axis")  
plt.show()  
```

```python
import matplotlib.pyplot as plt  

# Data for bar plot
categories = ['A', 'B', 'C', 'D']  
values = [4, 7, 1, 8]  

# Plotting the bar chart
plt.bar(categories, values)  
plt.title("Bar Plot")  
plt.xlabel("Categories")  
plt.ylabel("Values")  
plt.show()  
```

```python
import matplotlib.pyplot as plt  

# Data for histogram
data = [1, 2, 2, 3, 3, 3, 4, 5, 5, 5, 5]  

# Plotting the histogram
plt.hist(data, bins=5, alpha=0.7)  
plt.title("Histogram")  
plt.xlabel("Value")  
plt.ylabel("Frequency")  
plt.show()  
```

```python
import matplotlib.pyplot as plt  

# Data for scatter plot
x = [5, 7, 8, 5, 6]  
y = [7, 8, 9, 6, 5]  

# Plotting the scatter plot
plt.scatter(x, y)  
plt.title("Scatter Plot")  
plt.xlabel("X-axis")  
plt.ylabel("Y-axis")  
plt.show()  
```

```python
import numpy as np  
import matplotlib.pyplot as plt  

# Data for box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]  

# Plotting the box plot
plt.boxplot(data, vert=True, patch_artist=True)  
plt.title("Box Plot")  
plt.xlabel("Distribution")  
plt.ylabel("Values")  
plt.xticks([1, 2, 3], ['Std 1', 'Std 2', 'Std 3'])  
plt.show()  
```

```python
import numpy as np  
import matplotlib.pyplot as plt  

# Data for area plot
x = np.arange(1, 6)  
y1 = [1, 2, 3, 4, 5]  
y2 = [2, 3, 4, 5, 6]  

# Plotting the area plot
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4)  
plt.plot(x, y1, color='Slateblue', alpha=0.6, linewidth=2)  
plt.plot(x, y2, color='Slateblue', alpha=0.6, linewidth=2)  
plt.title("Area Plot")  
plt.xlabel("X-axis")  
plt.ylabel("Y-axis")  
plt.show()  
```

```python
import seaborn as sns  
import numpy as np  
import matplotlib.pyplot as plt  

# Data for heatmap
data = np.random.rand(10, 12)  

# Plotting the heatmap
sns.heatmap(data, annot=True)  
plt.title("Heatmap")  
plt.show()  
```

```python
import seaborn as sns  
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  

# Data for pair plot
df = pd.DataFrame({  
    'A': np.random.rand(50),  
    'B': np.random.rand(50),  
    'C': np.random.rand(50)  
})  

# Plotting the pair plot
sns.pairplot(df)  
plt.suptitle("Pair Plot")  
plt.show()  
```

```python
import seaborn as sns  
import matplotlib.pyplot as plt  

# Load example dataset
data = sns.load_dataset("tips")  

# Plotting the violin plot
sns.violinplot(x='day', y='total_bill', data=data)  
plt.title("Violin Plot")  
plt.show()  
```

```python
import matplotlib.pyplot as plt  
import numpy as np  

# Data for subplots
x = np.linspace(0, 10, 100)  
y1 = np.sin(x)  
y2 = np.cos(x)  

# Creating subplots
fig, axs = plt.subplots(2, 1)  
axs[0].plot(x, y1, 'r')  
axs[0].set_title('Sine Wave')  
axs[1].plot(x, y2, 'b')  
axs[1].set_title('Cosine Wave')  
plt.tight_layout()  
plt.show()  
```
