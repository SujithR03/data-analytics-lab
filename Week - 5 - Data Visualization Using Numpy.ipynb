## Numpy Histogram

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
data = np.random.randn(1000)

# Create histogram
plt.hist(data, bins=30, alpha=0.5, color='blue')

# Add title and labels
plt.title("Histogram from Numpy Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

## Numpy Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)

# Create scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title("Scatter Plot from Numpy Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

## Numpy Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for x and y
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create line plot
plt.plot(x, y)

# Add title and labels
plt.title("Line Plot from Numpy Data")
plt.xlabel("X-axis")
plt.ylabel("Sine Values")

# Show the plot
plt.show()
```
